```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
from get_aq_data import get_flo_data, ID_to_name
```

```python
data, hourly_mean, daily_mean = get_flo_data()
```

```python
subset = data['2019-04-15':'2019-05-30']
subset = subset.drop('nesta-8', axis=1)
```

```python
subset.apply(lambda x:)
```

```python
df = pd.DataFrame({'Max': subset.max(axis=0), '95th': subset.quantile(0.95, axis=0), '90th':subset.quantile(0.90, axis=0)})
df
```

```python
df.rename(index=ID_to_name).plot(grid=True, figsize=(10,6))
plt.suptitle('')
plt.title('Sensor comparison - 15th April to 30th May')
plt.ylabel(r'$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.xlabel('Sensor')
plt.tight_layout()
plt.savefig('graphs/CentilesAndMax.png', dpi=300)
```
