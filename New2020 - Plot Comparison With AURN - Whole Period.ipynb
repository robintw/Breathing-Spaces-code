```python
import importAURN

SOUT = importAURN.importAURN('SOUT', range(2018, 2022))

AURN = SOUT['2019-01-01':'2020-08-01']['PM2.5']
```

```python
all_site_daily_mean = daily_mean.mean(axis=1)
all_site_daily_std = daily_mean.std(axis=1)
```

```python
AURN_daily_mean = AURN.resample('1D').mean()
```

```python
ax = all_site_daily_mean['2019-01-01':].plot(figsize=(10, 6), label='Breathing Spaces')
AURN_daily_mean['2019-01-01':].plot(ax=ax, label="AURN")
plt.legend()
plt.grid(axis='both', which='major')
plt.xlabel('Time')
plt.ylabel('$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.title('Daily mean $\mathrm{PM}_{2.5}$: AURN vs Breathing Spaces')
plt.title('27th Feb 2019-29th Jun 2020', loc='right', fontstyle='italic')
plt.tight_layout()
plt.savefig('graphs2020/AURN_Comparison_WholePeriod.png', dpi=300)
```
