```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
from get_aq_data import get_flo_data, ID_to_name
```

```python
data, hourly_mean, daily_mean = get_flo_data()
```

```python
subset = data[:'2019-08-22'].mean(axis=1).resample('1H').mean()
#subset = data['2019-04-15':'2019-05-30'].mean(axis=1)
```

```python
subset = subset.to_frame()
```

```python
subset.columns = ['pm25']
```

```python
subset['date'] = subset.index.date
subset['hour'] = subset.index.hour
```

```python
reshaped = subset.pivot(columns='date', index='hour')
```

```python
reshaped = reshaped.T
```

```python
reshaped = reshaped.loc['pm25']
```

```python
reshaped
```

```python
reshaped.max(axis=1).index
```

```python
scaled = reshaped.values / reshaped.max(axis=1).values.reshape((169,1))
```

```python
scaled
```

```python
from sklearn.cluster import KMeans, DBSCAN, OPTICS, AgglomerativeClustering
```

```python
km = KMeans(n_clusters=20)
```

```python
km.fit(scaled)
```

```python
db = AgglomerativeClustering(n_clusters=15)
db.fit(scaled)
```
