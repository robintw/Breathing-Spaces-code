```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
from get_aq_data import get_flo_data, ID_to_name, get_all_sensor_data
```

```python
data, hourly_mean, daily_mean = get_all_sensor_data()
```

```python
ID_to_name
```

```python
hourly_mean = hourly_mean.drop(columns='nesta-4')
data = data.drop(columns='nesta-4')
```

```python
hourly_mean['2019-10-16':'2019-10-23'].rename(columns=ID_to_name).plot(figsize=(10, 6))
#plt.grid(axis='both', which='minor')
plt.grid(axis='both', which='major')
plt.xlabel('Time')
plt.ylabel('$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.title('Hourly average $\mathrm{PM}_{2.5}$')
plt.title('16th Oct-23rd Oct 2019', loc='right', fontstyle='italic')
plt.tight_layout()
plt.savefig('graphs/AllSites_Snapshot_WeekInOct.png', dpi=300)
```
