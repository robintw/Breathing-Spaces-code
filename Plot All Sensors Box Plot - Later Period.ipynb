```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from get_aq_data import get_flo_data, ID_to_name, annotate_boxplot
```

```python
data, hourly_mean, daily_mean = get_flo_data()
```

```python
subset = data['2019-07-14':]
subset = subset.drop(['nesta-8', 'nesta-5', 'nesta-1'], axis=1)
```

```python
bpdict = subset.rename(columns=ID_to_name).boxplot(showfliers=False, figsize=(10,6), whis=[5, 95], return_type='dict')
plt.suptitle('')
plt.title('Sensor comparison')
plt.title('14th Jul-23rd Aug 2019', loc='right', fontstyle='italic')
plt.ylabel(r'$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.xlabel('Sensor')
plt.tight_layout()
plt.savefig('graphs/SensorsBoxplot_JulOnwards.png', dpi=300)
annotate_boxplot(bpdict)
plt.annotate("Note: Percentage labels mean X% of sensor readings are below this value",
              (0.07, 0.015), xycoords='figure fraction', fontstyle='italic')
plt.tight_layout()
plt.savefig('graphs/SensorsBoxplot_JulOnwards_Ann.png', dpi=300)
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=subset.rename(columns=ID_to_name), cut=0, inner=None)
plt.grid()
plt.xlabel('')
plt.title('Sensor comparison')
plt.title('14th Jul-23rd Aug 2019', loc='right', fontstyle='italic')
plt.ylabel(r'$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.tight_layout()
plt.savefig('graphs/Sensors_ViolinPlot_JulOnwards.png', dpi=300)
plt.ylim(0, 10)
plt.savefig('graphs/Sensors_ViolinPlot_JulOnwards_Zoomed.png', dpi=300)
```
