```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
from get_aq_data import get_flo_data, ID_to_name
```

```python
data, hourly_mean, daily_mean = get_flo_data()
```

```python
ID_to_name
```

```python
subset = data['2019-07-14':]
subset = subset.drop(['nesta-8', 'nesta-5', 'nesta-1'], axis=1)
```

```python
subset.rename(columns=ID_to_name).boxplot(showfliers=False, figsize=(10,6))
plt.suptitle('')
plt.title('Sensor comparison - 14th July-23rd August')
plt.ylabel(r'$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.xlabel('Sensor')
plt.tight_layout()
plt.savefig('graphs/SensorsBoxplot_JulOnwards.png', dpi=300)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=subset.rename(columns=ID_to_name), cut=0, inner=None)
plt.grid()
plt.xlabel('')
plt.title('Sensor comparison - 14th July-23rd August')
plt.ylabel(r'$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.ylim(0, 10)
plt.tight_layout()
plt.savefig('graphs/Sensors_ViolinPlot_JulOnwards.png', dpi=300)
```
