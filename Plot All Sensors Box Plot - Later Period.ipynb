```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
from get_aq_data import get_flo_data, ID_to_name
```

```python
data, hourly_mean, daily_mean = get_flo_data()
```

```python
subset = data['2019-07-14':]
subset = subset.drop(['nesta-8', 'nesta-5', 'nesta-1'], axis=1)
```

```python
subset.rename(columns=ID_to_name).boxplot(showfliers=True, figsize=(10,6), whis=[5, 95])
plt.suptitle('')
plt.title('Sensor comparison - 14th July-23rd August')
plt.ylabel(r'$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.xlabel('Sensor')
plt.savefig('graphs/SensorsBoxplot_JulOnwards_WithOutliers.png', dpi=300)
# plt.annotate('Median\n(average)', (0.9, 3.8), xytext=(0.8, 4.5), arrowprops={'arrowstyle':'->'})
# plt.annotate('75%', (1.1, 5.8), xytext=(1.1, 6.5), arrowprops={'arrowstyle':'->'})
# plt.annotate('25%', (1.1, 2.6), xytext=(1.07, 3.3), arrowprops={'arrowstyle':'->'})
# plt.annotate('95%', (0.9, 10.77), xytext=(0.9, 11.3), arrowprops={'arrowstyle':'->'})
# plt.annotate('5%', (0.9, 1.77), xytext=(0.7, 2.2), arrowprops={'arrowstyle':'->'})
# plt.annotate("Note: percentage labels mean 'X% of values are below this level'",
#              (0.07, 0.03), xycoords='figure fraction', fontstyle='italic')
# plt.tight_layout()
# plt.savefig('graphs/SensorsBoxplot_JulOnwards_Ann.png', dpi=300)
```

```python
subset.rename(columns=ID_to_name).boxplot(showfliers=False, figsize=(10,6), whis=[5, 95])
plt.suptitle('')
plt.title('Sensor comparison - 14th July-23rd August')
plt.ylabel(r'$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.xlabel('Sensor')
plt.tight_layout()
plt.savefig('graphs/SensorsBoxplot_JulOnwards.png', dpi=300)
plt.annotate('Median\n(average)', (0.9, 3.8), xytext=(0.8, 4.5), arrowprops={'arrowstyle':'->'})
plt.annotate('75% less than this', (1.1, 5.8), xytext=(1.1, 6.5), arrowprops={'arrowstyle':'->'})
plt.annotate('25% less than this', (1.1, 2.65), xytext=(1.09, 3.3), arrowprops={'arrowstyle':'->'})
plt.annotate('95% less than this', (0.9, 10.77), xytext=(0.9, 11.3), arrowprops={'arrowstyle':'->'})
plt.annotate('5% less than this', (0.9, 1.77), xytext=(0.7, 2.2), arrowprops={'arrowstyle':'->'})
plt.tight_layout()
plt.savefig('graphs/SensorsBoxplot_JulOnwards_Ann2.png', dpi=300)
```
