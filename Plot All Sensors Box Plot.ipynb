```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
from get_aq_data import get_flo_data, ID_to_name, annotate_boxplot
```

```python
data, hourly_mean, daily_mean = get_flo_data()
```

```python
subset = data['2019-04-15':'2019-05-30']
subset = subset.drop('nesta-8', axis=1)
```

```python
bpdict = subset.rename(columns=ID_to_name).boxplot(showfliers=False, figsize=(10,6), return_type='dict')
plt.suptitle('')
plt.title('Sensor comparison')
plt.title('15th Apr-30th May 2019', loc='right', fontstyle='italic')
plt.ylabel(r'$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.xlabel('Sensor')
plt.tight_layout()
plt.savefig('graphs/SensorsBoxplot.png', dpi=300)
annotate_boxplot(bpdict)
plt.annotate("Note: Percentage labels mean X% of sensor readings are below this value",
              (0.07, 0.010), xycoords='figure fraction', fontstyle='italic')
plt.savefig('graphs/SensorsBoxplot_Ann.png', dpi=300)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=subset.rename(columns=ID_to_name), cut=0, inner=None)
plt.grid()
plt.xlabel('')
plt.title('Sensor comparison')
plt.title('15th Apr-30th May 2019', loc='right', fontstyle='italic')
plt.ylabel(r'$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.tight_layout()
plt.savefig('graphs/Sensors_ViolinPlot.png', dpi=300)
plt.annotate('Lots of\ndata\npoints', xy=(0.3, 6), xytext=(0.3, 12), arrowprops={'arrowstyle':'->'})
plt.annotate('Few\ndata\npoints', xy=(0.05, 25), xytext=(0.3, 28), arrowprops={'arrowstyle':'->'})
plt.annotate('Maximum', xy=(0, 71), xytext=(0.2, 75), arrowprops={'arrowstyle':'->'})
plt.annotate('Minimum', xy=(0, -0.5), xytext=(0.3, -3), arrowprops={'arrowstyle':'->'})
plt.tight_layout()
plt.savefig('graphs/Sensors_ViolinPlot_Ann.png', dpi=300)
```
