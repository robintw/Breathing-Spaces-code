```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from get_aq_data import get_flo_data, ID_to_name, TICKS_TWO_HOURLY
```

```python
data, hourly_mean, daily_mean = get_flo_data()
```

```python
subset = data['2019-05-05'].mean(axis=1)
```

```python
subset.plot(figsize=(10, 6))
plt.grid(axis='both', which='minor')
plt.grid(axis='both', which='major')
plt.fill_betweenx(np.arange(3, 11), pd.to_datetime('2019-05-05 08:00:00'), pd.to_datetime('2019-05-05 16:00:00'), color='red', alpha=0.2)
plt.ylim(3, 10)
plt.suptitle('')
plt.annotate('Road closure period', xy=(pd.to_datetime('2019-05-05 12:00:00'), 9.3), ha='center', fontstyle='italic')
plt.title('Average $\mathrm{PM}_{2.5}$ time-series on marathon day')
plt.ylabel(r'$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.xlabel('Hour')
plt.tight_layout()
plt.savefig('graphs/MarathonTimeSeries.png', dpi=300)
```
