```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
from get_aq_data import get_flo_data, ID_to_name
```

```python
data, hourly_mean, daily_mean = get_flo_data()
```

```python
subset = data
subset = subset.drop('nesta-8', axis=1)
subset = subset.mean(axis=1)
just_sundays = subset[subset.index.dayofweek == 6].to_frame('Mean')
```

```python
just_sundays['Marathon'] = 'Non-Marathon'
just_sundays.loc['2019-05-05', 'Marathon'] = 'Marathon'
```

```python
just_sundays.boxplot(by='Marathon', figsize=(10, 6), whis=[5, 95])
plt.suptitle('')
plt.title('$\mathrm{PM}_{2.5}$ on marathon day (5th May 2019) vs other Sundays')
plt.title('7th Mar-23rd Aug 2019', loc='right', fontstyle='italic')
plt.ylabel(r'$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.xlabel('Sensor')
plt.tight_layout()
plt.savefig('graphs/MarathonBoxplot_WholePeriod.png', dpi=300)
```

```python
just_sundays.boxplot(by='Marathon', figsize=(10, 6), showfliers=False, whis=[5, 95])
plt.suptitle('')
plt.title('$\mathrm{PM}_{2.5}$ on marathon day (5th May 2019) vs other Sundays')
plt.title('7th Mar-23rd Aug 2019', loc='right', fontstyle='italic')
plt.ylabel(r'$\mathrm{PM}_{2.5}$ ($\mu g / m^3$)')
plt.tight_layout()
plt.savefig('graphs/MarathonBoxplot_WholePeriod_NoOutliers.png', dpi=300)
```
